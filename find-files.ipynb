```python
import pandas as pd
from pathlib import Path
RANDOM_SEED = 42 
```

Random samples of the CMIP5 data

```python
cmip5_file_names = pd.read_csv('data/cmip5_files.txt', header=None)
cmip5_file_names['ext'] = cmip5_file_names[0].map(lambda x: Path(x).suffix)
cmip5_file_names = cmip5_file_names.loc[cmip5_file_names.ext == '.nc']

cmip5_subset = cmip5_file_names.sample(300, random_state=RANDOM_SEED)

cmip5_subset[0] = cmip5_subset[0].map(lambda x: '/badc/cmip5/data/' + x)
cmip5_subset.drop('ext', axis=1).to_csv('data/cmip5_files_subset.txt', index=None)
cmip5_subset
```

Random samples of the CRU data

```python
cru_file_names = pd.read_csv('data/cru_files.txt', header=None)

cru_file_names['ext'] = cru_file_names[0].map(lambda x: Path(x).suffix)
cru_file_names = cru_file_names.loc[cru_file_names.ext == '.gz']

cru_subset = cru_file_names.sample(300, random_state=RANDOM_SEED)

cru_subset[0] = cru_subset[0].map(lambda x: '/badc/cru/' + x)
cru_subset.drop('ext', axis=1).to_csv('data/cru_files_subset.txt', index=None)
cru_subset
```
